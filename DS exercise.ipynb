```python
pid = 'Package ID'
gluten = 'Gluten Content (%)'
elasticity = 'Dough Elasticity Index'
dampening = 'Dampening Time (hours)'
weight = 'Package Weight (g)'
ash = 'Ash content (%)'
recipe = 'Production Recipe'
moisture = 'Moisture (%)'
starch = 'Starch Content (%)'
mill = 'Production Mill'
volume = 'Package Volume (cm3)'
proteins = 'Proteins (g)/100g'
color = 'Color'
quality = 'Quality'
unnamed = 'Unnamed: 14'
```

```python
df_filtered = df[df.isnull().sum(1) == 1].drop(columns=[unnamed])
```

```python
from matplotlib import pyplot as plt

def plot_all(dflist, **kwargs):

    opts = {'alpha':0.5, 'histtype':'stepfilled', 'label':None}
    opts.update(kwargs)
    def sanitize_filename(f):
        return "".join(x for x in f if x.isalnum())
    dfref = dflist[0]

    rows = dfref.shape[1]//3+1
    fig, axs = plt.subplots(rows,3, figsize=(15, rows*5))
    axs = axs.reshape(-1)

    for i, col in enumerate(dfref.columns):
        ax = axs[i]
        ax.set_title(col)
        ax.hist([df[col] for df in dflist], **opts)
        ax.set_yscale("log")
        #_, limmax = ax.get_ylim()
        #ax.set_ylim(0.5, limmax)
        if opts['label']:
            ax.legend()
    fig.savefig(f"{sanitize_filename(col)}.png")
```

```python
plot_all([df_filtered], bins=100)
```

```python
outliers = (df_filtered[ash] > 10) | (df_filtered[dampening] < 0) | (df_filtered[elasticity] > 100) | (df_filtered[gluten] < 0)
df_good = df_filtered[~outliers]
df_outliers = df_filtered[outliers]
```

```python
plot_all([df_outliers,df_good],label=["outliers","good"])
```
